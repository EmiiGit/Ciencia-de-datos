#Detección de outliers

```python
# Configurar gráficos
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

# Cargar datos (ajustá el nombre de tu archivo si es diferente)
campaigns = pd.read_excel('datos_BI_transformados.xlsx', sheet_name='Campaigns_all')

# Calcular ROAS
campaigns['ROAS'] = campaigns['revenue'] / campaigns['spend']

print("=" * 60)
print("DETECCIÓN DE OUTLIERS")
print("=" * 60)

def detectar_outliers(df, columna):
    Q1 = df[columna].quantile(0.25)
    Q3 = df[columna].quantile(0.75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR
    outliers = df[(df[columna] < limite_inferior) | (df[columna] > limite_superior)]
    return outliers, limite_inferior, limite_superior

# 1. BOXPLOTS
metricas_outliers = ['spend', 'revenue', 'impressions', 'clicks', 'purchase']
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
for i, metrica in enumerate(metricas_outliers):
    row, col = i // 3, i % 3
    data = campaigns[campaigns[metrica] > 0][metrica]
    axes[row, col].boxplot(data, vert=True, patch_artist=True)
    axes[row, col].set_title(f'Boxplot - {metrica}', fontsize=12)
    axes[row, col].set_ylabel(metrica)
    axes[row, col].set_yscale('log')
axes[1, 2].axis('off')
plt.suptitle('Detección de Outliers por Métrica (Escala Log)', fontsize=14)
plt.tight_layout()
plt.show()

# 2. GRÁFICO DE DISPERSIÓN con outliers en rojo
# Identificar outliers en spend
outliers_spend, _, _ = detectar_outliers(campaigns[campaigns['spend'] > 0], 'spend')

plt.figure(figsize=(14, 8))

# Campañas normales
normal = campaigns[~campaigns['campaign_id'].isin(outliers_spend['campaign_id']) & (campaigns['spend'] > 0)]
plt.scatter(normal['spend'], normal['ROAS'], alpha=0.5, c='lightblue', label='Campañas normales', s=80)

# Outliers (rojo)
outliers_plot = campaigns[campaigns['campaign_id'].isin(outliers_spend['campaign_id'])].copy()
plt.scatter(outliers_plot['spend'], outliers_plot['ROAS'], alpha=0.8, c='red', label='Outliers', s=200)

# Etiquetas para los outliers
for _, row in outliers_plot.iterrows():
    plt.annotate(row['campaign_id'], (row['spend'], row['ROAS']),
                 xytext=(5, 5), textcoords='offset points', fontsize=9, fontweight='bold')

plt.axhline(y=1, color='red', linestyle='--', label='ROAS = 1')
plt.xscale('log')
plt.xlabel('Spend (log scale)')
plt.ylabel('ROAS')
plt.title('Dispersión Spend vs ROAS - Outliers en rojo', fontsize=14)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Calcular CPA por región, objetivo y canal
cpa_analysis = campaigns_valid.groupby(['region', 'objective', 'media_type']).agg({
    'spend': 'sum',
    'purchase': 'sum'
}).reset_index()
cpa_analysis['CPA'] = cpa_analysis['spend'] / cpa_analysis['purchase']

# Filtrar solo CPA válidos (purchase > 0)
cpa_analysis = cpa_analysis[cpa_analysis['purchase'] > 0]

print("📊 TABLA CPA POR REGIÓN + OBJETIVO + CANAL:")
print(cpa_analysis[['region', 'objective', 'media_type', 'CPA']].to_string(index=False))

# Pivot para tener columnas separadas
pivot_cpa = cpa_analysis.pivot_table(index=['region', 'media_type'], columns='objective', values='CPA').fillna(0)

# Gráfico de barras agrupadas
fig, ax = plt.subplots(figsize=(14, 8))
x = np.arange(len(pivot_cpa.index))
width = 0.25

# Crear grupos de barras
for i, obj in enumerate(['awareness', 'consideration', 'conversion']):
    if obj in pivot_cpa.columns:
        values = pivot_cpa[obj].values
        offset = (i - 1) * width
        bars = ax.bar(x + offset, values, width, label=obj.capitalize())

        # Agregar etiquetas de valores (solo si > 0)
        for bar in bars:
            height = bar.get_height()
            if height > 0:
                ax.annotate(f'${height:,.0f}', xy=(bar.get_x() + bar.get_width()/2, height),
                           xytext=(0, 3), textcoords="offset points", ha='center', va='bottom', fontsize=7, rotation=90)

ax.set_xticks(x)
ax.set_xticklabels([f"{r[0]}\n{r[1]}" for r in pivot_cpa.index], rotation=45, ha='right')
ax.set_ylabel('CPA ($)')
ax.set_title('CPA por Región, Canal y Objetivo', fontsize=14)
ax.axhline(y=cpa_analysis['CPA'].mean(), color='red', linestyle='--', label=f'Promedio Global: ${cpa_analysis["CPA"].mean():,.0f}')
ax.legend()
plt.tight_layout()
plt.show()

# Destacar la mejor y peor combinación
mejor = cpa_analysis.loc[cpa_analysis['CPA'].idxmin()]
peor = cpa_analysis.loc[cpa_analysis['CPA'].idxmax()]

print("\n" + "=" * 60)
print("🎯 MEJOR COMBINACIÓN (menor CPA):")
print(f"   {mejor['region']} | {mejor['objective']} | {mejor['media_type']} → ${mejor['CPA']:,.0f}")

print("\n💀 PEOR COMBINACIÓN (mayor CPA):")
print(f"   {peor['region']} | {peor['objective']} | {peor['media_type']} → ${peor['CPA']:,.0f}")

print(f"\n📊 Diferencia: {peor['CPA']/mejor['CPA']:.1f}x más caro")
```
